```python
# we use the describe values to create a function which will help put the row values in total claim amount to different bins
def c_bins(row):
    if row < 272:
        return ' < 25%'
    elif row < 383:
        return '25%-50%'
    elif row < 548:
        return '50%-75%'
    else:
        return '> 75%'
# go and get every row of total claim amount and apply the bins() function to every row
data['claim bin'] = data['total_claim_amount'].apply(lambda row: c_bins(row))
data.head()
```

```python
X_num.describe()
```

```python
# NORMALIZE WITH STANDARDSCALER
from sklearn.preprocessing import StandardScaler

X_scaled=StandardScaler().fit_transform(X_num)

def normalize(X):
    X_mean=X.mean(axis=0)
    X_std=X.std(axis=0)
    X_std[X_std==0]=1.0
    X=(X-X_mean)/X_std
    return X

X_num=normalize(X_num)
X_num.head()
```

```python
(X_num-X_scaled).sum().sum() 
```

```python
# CATEGORICAL
pd.get_dummies(X_cat, drop_first=True).head()
```

```python
# CONCAT DATAFRAMES
X_cat=pd.get_dummies(X_cat, drop_first=True)

# CONCAT NUMERICAL AND CATEGORICAL TRANSFORMATIONS
X=pd.concat([X_num, X_cat], axis=1) 
X.head()
```

LINEAR REGRESSION

```python
# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split as tts
import warnings
warnings.simplefilter('ignore')
import statsmodels.api as sm
from statsmodels.formula.api import ols
from sklearn.linear_model import LinearRegression as LinReg
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
```

```python
X_train, X_test, y_train, y_test=tts(X, y, test_size=.2)
```

```python
X_train_wi=sm.add_constant(X_train)
model=sm.OLS(y_train,X_train_wi).fit()

print(model.summary())
```

```python
#  lINEAR REGRESSION MODEL 
linreg=LinReg()    

# TRAIN MODEL
linreg.fit(X_train, y_train) 

# PREDICTION MODEL
y_pred_linreg=linreg.predict(X_test)  
```

MODEL VALIDATION

```python
# R2
print ('train R2: {} -- test R2: {}'.format(linreg.score(X_train, y_train),
                                            linreg.score(X_test, y_test)))
```

```python
# MSE
train_mse=mse(linreg.predict(X_train), y_train)
test_mse=mse(linreg.predict(X_test), y_test)
```

```python
print ('train MSE: {} -- test MSE: {}'.format(train_mse, test_mse))
```

```python
# RMSE
print ('train RMSE: {} -- test RMSE: {}'.format(train_mse**.5, test_mse**.5))
```

```python
# MAE
train_mae=mae(linreg.predict(X_train), y_train)
test_mae=mae(linreg.predict(X_test), y_test)
```

```python
print ('train MAE: {} -- test MAE: {}'.format(train_mse, test_mse))
```
